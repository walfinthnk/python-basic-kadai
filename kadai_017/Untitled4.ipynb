```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} is an adult.")
        else:
            print(f"{self.name} is not an adult.")

taro = Human("Taro", 15)
hanako = Human("Hanako", 25)
jiro = Human("Jiro", 18)
keiko = Human("Keiko", 32)

humans = [taro, hanako, jiro, keiko]

for human in humans:
    human.check_adult()


```
